```python
# version without splitting

import numpy as np
import matplotlib.pyplot as plt
from matplotlib.animation import FuncAnimation
from IPython.display import HTML
from scipy.fftpack import fft2, ifft2
from scipy.integrate import solve_ivp

#define parameters
tspan = np.arange(0, 4.1, 0.5)
Lx, Ly = 20, 20
nx, ny = 64, 64
N = nx * ny
D1, D2 = 0.1, 0.1
beta = 1

#spacial domain
x2 = np.linspace(-Lx/2, Lx/2, nx + 1)
x = x2[:nx]
y2 = np.linspace(-Ly/2, Ly/2, ny + 1)
y = y2[:ny]
dx = np.abs(x[1] - x[0])
dy = np.abs(y[1] - y[0])
deltaSqr = dx**2

#k values
kx = (2 * np.pi / Lx) * np.concatenate((np.arange(0, nx/2), np.arange(-nx/2, 0)))
kx[0] = 1e-6
ky = (2 * np.pi / Ly) * np.concatenate((np.arange(0, ny/2), np.arange(-ny/2, 0)))
ky[0] = 1e-6
KX, KY = np.meshgrid(kx, ky)
K =  KX**2 + KY**2

#initial conditions
X, Y = np.meshgrid(x, y)
m = 1   # number of spirals
u = np.tanh(np.sqrt(X**2+Y**2))*np.cos(m*np.angle(X+1j*Y)-(np.sqrt(X**2+Y**2))) + 1j * np.zeros((nx, ny))
v = np.tanh(np.sqrt(X**2+Y**2))*np.sin(m*np.angle(X+1j*Y)-(np.sqrt(X**2+Y**2))) + 1j * np.zeros((nx, ny))

#ut0 = np.hstack([np.real(fft2(u).reshape(N)), np.imag(fft2(u).reshape(N))])
#vt0 = np.hstack([np.real(fft2(v).reshape(N)), np.imag(fft2(v).reshape(N))])
#qt0 = np.hstack([ut0, vt0])
qt0 = np.hstack([fft2(u).reshape(N), fft2(v).reshape(N)])

#theode
def spct_rhs2(t, qt, D1, D2, beta, K, nx, ny, N):
    #reshape    qt shape [4N,]  : u[real]/u[imag]/v[real]/v[imag]
    ut_1D = qt[0:N]
    vt_1D = qt[N:2*N]
    ut_2D = ut_1D.reshape((nx,ny))
    vt_2D = vt_1D.reshape((nx,ny))
    u_2D = ifft2(ut_2D)
    v_2D = ifft2(vt_2D)
    # calculate "forward"
    A2 = u_2D**2 + v_2D**2
    lamA  = 1 - A2
    omegA = -beta*A2
    utnew_2D = fft2(lamA*u_2D - omegA*v_2D) + (D1*-K*ut_2D)
    vtnew_2D = fft2(omegA*u_2D + lamA*v_2D) + (D2*-K*vt_2D)
    #reshape
    utnew_1D = utnew_2D.reshape(N)
    vtnew_1D = vtnew_2D.reshape(N)
    return np.hstack([utnew_1D, vtnew_1D])


solt = solve_ivp(spct_rhs2, (tspan[0], tspan[-1]), qt0, args = (D1, D2, beta, K, nx, ny, N), t_eval=tspan, method='RK45',)
A1 = solt.y

display(A1)
display(A1.shape)

# for j, t in enumerate(tspan):
#     plt.figure()
#     plt.pcolor(x, y, np.real(ifft2((A1[:N,j]+1j*A1[N:2*N,j]).reshape((nx, ny)))), cmap = 'rainbow', shading='auto')
    

################################################~ANIMATE~###########################################################

# def update(k, A1, N):
#     pcm.set_array(np.real(ifft2((A1[:N,k]+1j*A1[N:2*N,k]).reshape((nx, ny)))))  # Set the data for the new frame
#     return pcm,

# fig, ax = plt.subplots(figsize=(8,6))
# pcm = ax.pcolormesh(x, y, np.real(ifft2((A1[:N,0]+1j*A1[N:2*N,0]).reshape((nx, ny)))), cmap='rainbow', shading='auto')
# fig.colorbar(pcm)
# plt.close(fig)
# # Create the animation object
# ani = FuncAnimation(fig, update, frames=len(tspan) - 1, fargs=(A1, N), interval=80, blit=False)

# #animation
# HTML(ani.to_jshtml())
```

```python
import numpy as np
from numpy import *
import matplotlib.pyplot as plt
from scipy.integrate import solve_ivp
from matplotlib.animation import FuncAnimation
from IPython.display import HTML

def cheb(N):
	if N==0: 
		D = 0.; x = 1.
	else:
		n = arange(0,N+1)
		x = cos(pi*n/N).reshape(N+1,1) 
		c = (hstack(( [2.], ones(N-1), [2.]))*(-1)**n).reshape(N+1,1)
		X = tile(x,(1,N+1))
		dX = X - X.T
		D = dot(c,1./c.T)/(dX+eye(N+1))
		D -= diag(sum(D.T,axis=0))
	return D, x.reshape(N+1)

#define parameters
tspan = np.arange(0, 4.1, 0.5)
D_1, D_2 = 0.1, 0.1
beta = 1
N = 30
D, x = cheb(N)
D = (1/10)*D  # <-scale factor to get out of -1 to 1
D[N, :] = 0
D[0, :] = 0
D2 = np.dot(D, D)
I = np.eye(len(D2))
L = kron(I, D2) + kron(D2, I) 
x = 10*x     # <- scale factor to get out of -1 to 1
y = x

#initial conditions
X, Y = np.meshgrid(x, y)
m = 1   # number of spirals
u = np.tanh(np.sqrt(X**2+Y**2))*np.cos(m*np.angle(X+1j*Y)-(np.sqrt(X**2+Y**2)))
v = np.tanh(np.sqrt(X**2+Y**2))*np.sin(m*np.angle(X+1j*Y)-(np.sqrt(X**2+Y**2)))

#flatten
q0 = np.hstack([u.reshape((N+1)**2), v.reshape((N+1)**2)])

def spct_rhs(t, q, D1, D2, beta, L):
    u1 = q[0:(N+1)**2]
    v1 = q[(N+1)**2:]
    u2 = u1.reshape(N+1,N+1)
    v2 = v1.reshape(N+1,N+1)
    A2 = u2**2 + v2**2
    lamA  = 1 - A2
    omegA = -beta*A2
    unew = lamA*u2 - omegA*v2 + D1*(np.dot(L, u1).reshape(N+1, N+1))
    vnew = omegA*u2 + lamA*v2 + D2*(np.dot(L, v1).reshape(N+1, N+1))
    return np.hstack([unew.reshape((N+1)**2), vnew.reshape((N+1)**2)])


soln = solve_ivp(spct_rhs, (tspan[0], tspan[-1]), q0, args = (D_1, D_2, beta, L), t_eval=tspan, method='RK45',)

A2 = soln.y
display(A2)
display(A2.shape)

################################################~ANIMATE~###########################################################
# def update(k, w_storage, N):
#     pcm.set_array(A2[:(N+1)**2, k+1].reshape(N+1,N+1))  # Set the data for the new frame
#     return pcm,

# # Plot setup
# fig, ax = plt.subplots(figsize=(8, 6))
# pcm = ax.pcolormesh(x, y, A2[:(N+1)**2,0].reshape(N+1,N+1), cmap='rainbow', shading='auto')
# fig.colorbar(pcm)
# plt.close(fig)
# # Create the animation object
# ani = FuncAnimation(fig, update, frames=len(tspan) - 1, fargs=(A2, N), interval=100, blit=False)

# #animation
# HTML(ani.to_jshtml())
```
